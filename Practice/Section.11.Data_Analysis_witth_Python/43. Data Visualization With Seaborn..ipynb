## Data Visualization with Seaborn
Seaborn is a Python visualization library based on Matplotlib that provides a high-level interface for drawing
attractive and informative statistical graphics. Seaborn helps in creating complex visualizations with just a few
lines of code. 

```python
import seaborn as sns
```

```python
#basic ploting with seaborn
tips=sns.load_dataset('tips')
tips.head()
```

```python
#create a scatter plot
import matplotlib.pyplot as plt
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title('Total Bill vs Tip')
plt.show()
```

```python
#line plot
sns.lineplot(x='size',y='total_bill',data=tips)
plt.title('line plot ')
plt.show()
```

```python
#categorical plot
#bar plot
sns.barplot(x='day',y='total_bill',data=tips)
plt.title('Bar plot ')
plt.show()
```

```python
#box plot
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title('Box plot ')
plt.show()
```

```python
#violin plot
sns.violinplot(x='day',y='total_bill',data=tips)
plt.title('Violin plot ')
plt.show()
```

```python
#histogram
sns.histplot(tips['total_bill'],bins=10,kde=True)
plt.title('Histogram')
plt.show()
```

```python
##kde plot
sns.kdeplot(tips['total_bill'],fill=True)
plt.title('KDE plot')
plt.show()
```

```python
#pairplot
sns.pairplot(tips,hue='sex')
plt.show()
```

```python
#heatmap
corr= tips[['total_bill','tip','size']].corr()
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.show()
```

```python
import pandas as pd
sales_df=pd.read_csv('sales_data.csv')
sales_df.head()
```

```python
#plot total sales by product
plt.figure(figsize=(10,5))
sns.barplot(x='Product Category',y='Total Revenue',data=sales_df,estimator=sum)
plt.show()
```

```python
#plot total sales by region
plt.figure(figsize=(10,5))
sns.barplot(x='Region',y='Total Revenue',data=sales_df,estimator=sum)
plt.show()
```
