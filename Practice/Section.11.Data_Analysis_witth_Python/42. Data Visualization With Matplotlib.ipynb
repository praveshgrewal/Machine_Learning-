## Data Visualization with Matplotlib
is a powerful plotting library for python that enables the creation of static animated and interactive visualizations. It is widely used for data visualization in data science and analytics. in the lesson we will cover the basic of Matplotlib, including creating various types of plot and customizing them.

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
#create line graph
plt.plot(x,y)
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('Sample graph')
plt.show()


```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
#customized line plot
plt.plot(x,y,color='red',linestyle='-.',marker='o',linewidth=3,markersize=5)
plt.grid(True)
```

```python
#multiple  plot
#sample data
x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,3,5,7,9]
#customized line plot
plt.figure(figsize=(9,5))
plt.subplot(1,2,1) #one row, two columns, first plot
plt.plot(y1,x,color='red')
plt.title('First plot')

plt.subplot(1,2,2) #one row, two columns, second plot
plt.plot(x,y1,color='green')
plt.title('second plot')
```

```python
#bar plot
#categories
categories=['Frogs','Hogs','Dogs','Logs']
#values
values=[15, 30, 45, 10]
#create bar plot
plt.bar(categories,values,color='purple')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar plot')
plt.show()
```

## histogram
are used to represent the distribution of a dataset. they divide the data into bins and count the number of data points in each bin

```python
#histogram
#sample data
ages=[2,5,70,40,30,45,50,45,43,40,44,60,7,13,57,18,90,77,32,21,20,40]
#create histogram
plt.hist(ages,bins=5,color='orange',edgecolor='black')
```

```python
#create a scatter plot
x=[1,2,3,4,5]
y=[1,4,9,16,25]
plt.scatter(x,y,color='red',marker='o',s=50)
plt.show()
```

```python
#pie chart
#sample data
x=[1,2,3,4,5]
y=[1,4,9,16,25]
colors= ['red','green','blue','cyan','magenta']
explode=[0.3,0,0,0,0]
#create pie chart
plt.pie(x,labels=y,explode=explode,colors=colors,autopct='%1.1f%%',startangle=90,shadow=True)
plt.show()
```

```python
#sales data visualization
import pandas as pd

df=pd.read_csv('sales_data.csv')
df.head()
```

```python
#plot total sales by product
total_sales=df.groupby('Product Category')['Total Revenue'].sum()
print(total_sales)
```

```python
#bar plot
total_sales.plot(kind='bar',color='green')
plt.show()
```

```python
#plot sales trend over time
sales_trend=df.groupby('Date')['Total Revenue'].sum().reset_index()
plt.plot(sales_trend['Date'],sales_trend['Total Revenue'])
```
